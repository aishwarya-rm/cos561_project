```python
import scapy.all as sc
import matplotlib.pyplot as plt
from matplotlib import dates
import pandas as pd
import numpy as np
import datetime
```

```python
def make_plot(filename):
    pkts = sc.rdpcap(filename)
    times = [p.time for p in pkts]
    data = [p.len for p in pkts]
    datetimes = [datetime.datetime.fromtimestamp(c) - datetime.datetime.fromtimestamp(times[0]) for c in times]
    d = {'timestamp': datetimes, 'pkt_sizes': data}
    df = pd.DataFrame(data=d)
    df.groupby('timestamp')['pkt_sizes'].sum()
    plt.title(filename)
    plt.plot(df.groupby('timestamp')['pkt_sizes'].sum() / 1000 * 8)
    plt.ylabel('Video Flow Throughput (Kbps)')
    plt.xlabel('Time (minutes)')
    plt.savefig(filename+'.png')
    plt.show()
```

```python
for f in ['analyze_1204_4.cap', 'analyze_1204_2.cap', 'analyze_1204_1.cap', 'analyze_1204_.5.cap']:
    make_plot('data/' + f)
```
